```python
#operation and conditional statement :(if ,else,elif)


score = 8

if score >7:
    print("Above Average : Excellent Performance!")
elif 4 <= score <= 7:
    print("Average :good effort! keep practicing")
else:
    print("Below Average : Need to improve, consistent practice helps.")
```
